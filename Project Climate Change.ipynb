```python
df=df.dropna()
```

```python
df_selected=df[['Year', 'Avg Temperature (°C)', 'Rainfall (mm)']]
print(df_selected.head())

```

```python
df_yearly = df.groupby("Year").mean().reset_index()
df=df.sort_values('Year')
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Year'], df['Avg Temperature (°C)'], color="red")
plt.title("Temperature Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Temperature")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df_yearly['Year'], df_yearly['Rainfall (mm)'],
color="blue", marker="o", linewidth=2)
plt.title("Rainfall Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Rainfall")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df_yearly['Year'], df_yearly['Avg Temperature (°C)'],
label="Temperature", color="red", marker="o")
plt.plot(df_yearly['Year'], df_yearly['Rainfall (mm)'], 
label="Rainfall", color="blue", marker="o")
plt.title("Climate Change: Temperature vs Rainfall Trends")
plt.xlabel("Year")
plt.ylabel("Values")
plt.legend()
plt.grid(True)
plt.show()
```
